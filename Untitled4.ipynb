```python
#calculate the avg marks and print class failed or passed in avg is 50 or above

marks=[45,60,72,38,90]

total = 0
for i in range(len(marks) -1):
  total +=marks[i]
average = total / len(marks)
if average >50:
  print("Class passed")
else:
  print("Class failed")
print("Average",average)
```

```python
import pandas as pd
data = {
    "student_id":[101,102,103,104,105,106],
    "attendence_percentage":[92,67,81,45,74,88],
    "assignment_score":[18,12,15,8,14,19],
    "quiz_score":[72,48,65,30,55,80],
    "lab_completed":[True,False,True,False,True,True]
}
df = pd.DataFrame(data)
df
```

```python
df.shape
(6,5)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["total_score"]=df["assignment_score"]+df["quiz_score"]
df
```

```python
df["eligible"]=(
    (df["attendence_percentage"]>=75)&
    (df["total_score"]>=70)&
    (df["lab_completed"]==True)
)
df
```

```python
summary={
    "average_attendence":df["attendence_percentage"].mean(),
   "average_score":df["total_score"].mean(),
    "eligible_count":df["eligible"].sum(),
    "total_student":len(df)
}
summary
```

```python
assert df["attendence_percentage"].between(0,100).all()
assert df["assignment_score"].between(0,20).all()
assert df["quiz_score"].between(0,100).all()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))

plt.bar(df["student_id"],df["quiz_score"])

plt.title("Student Quiz Scores")
plt.xlabel("Student ID")
plt.ylabel("Quiz Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["student_id"],df["attendence_percentage"])
plt.title("Student Attendence")
plt.xlabel("Student ID")
plt.ylabel("Attendence Percentage")
plt.show()
```
